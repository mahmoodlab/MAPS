```python
import os
import numpy as np
import pandas as pd
import utils

from matplotlib import pyplot as plt
from cell_phenotyping import Trainer, Predictor
```

## Relative Paths 

```python
train_data_path = '../data/cell_phenotyping/train.csv'
valid_data_path = '../data/cell_phenotyping/valid.csv'
results_dir = '../results/cell_phenotyping/'

class_names = ['B', 'CD4 T', 'CD8 T', 'DC', 'Endothelial', 'Epithelial', 'Lymphatic', 'M1', 'M2', 'Mast', 'Monocyte', 'NK', 'Neutrophil', 'Other', 'Treg', 'Tumor']
color_palette =['#bed8d9', '#2bb8b2', '#fed53f', '#f98866', '#fe0000']
```

# Training a new cell phenotyping model from cell expression data

```python
model = Trainer(results_dir=results_dir, num_features=50, num_classes=16, batch_size=128, max_epochs=500, min_epochs=250, patience=100, verbose=0)
model.fit(train_data_path, valid_data_path)
pred_labels, pred_probs = model.predict(valid_data_path)

gt_labels = pd.read_csv(valid_data_path)['cell_label'].to_numpy()
utils.visulize_training_logs(os.path.join(results_dir, 'training_logs.csv'))
utils.conf_matrix(gt_labels, pred_labels, class_names[:])
utils.performance_plots(gt_labels, pred_labels, class_names[:], color_palette)
```

# Cell phenotype prediction on validation dataset using pretrained cell phenotyping model

```python
pretrained_model_checkpoint_path = '../models/cell_phenotyping/cHL_CODEX.pt'
model = Predictor(model_checkpoint_path=pretrained_model_checkpoint_path, num_features=50, num_classes=16, batch_size=128)

pred_labels, pred_probs = model.predict(valid_data_path)
gt_labels = pd.read_csv(valid_data_path)['cell_label'].to_numpy()
res_dict = {}
for i, class_name in enumerate(class_names):
    res_dict[class_name+'_prob'] = pred_probs[:,i]
res_dict['pred_label'] = pred_labels
res_dict['gt_label'] = gt_labels
pd.DataFrame.from_dict(res_dict).to_csv(os.path.join(results_dir, 'results_valid.csv'), index=False)

utils.conf_matrix(gt_labels, pred_labels, class_names[:])
utils.performance_plots(gt_labels, pred_labels, class_names[:], color_palette)
```
